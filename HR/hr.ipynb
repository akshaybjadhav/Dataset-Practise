```python
data = df.copy()
```

```python
from sklearn.model_selection import train_test_split
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from hyperopt import hp, fmin, tpe, STATUS_OK, Trials
```

```python
data.head(3)
```

```python
cat = ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 
       'JobRole', 'MaritalStatus', 'Over18', 'OverTime']
dummy = pd.get_dummies(data[cat],drop_first=True)
```

```python
data = pd.concat([dummy,data],axis=1)
data.head()
```

```python
data.drop(columns=cat,axis=1,inplace=True)
```

```python
data.head()
```

```python
data.rename(columns={'Attrition_Yes':'Attrition'},inplace=True)
data.head(2)
```

```python
x = data.drop(columns=['Attrition'],axis=1)
y = data['Attrition']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20)
```

```python
lc = LGBMClassifier()
lc.fit(x_train,y_train)
y_predict = lc.predict(x_test)
print(accuracy_score(y_test,y_predict))
print(confusion_matrix(y_test,y_predict))
print(classification_report(y_test,y_predict))
```

```python
# defining parameters for LGB
space = {
    'learning_rate': hp.loguniform('learning_rate', np.log(0.001), np.log(0.3)),
    'n_estimators': hp.quniform('n_estimators', 50, 1200, 25),
    'max_depth': hp.quniform('max_depth', 1, 15, 1),
    'num_leaves': hp.quniform('num_leaves', 10, 150, 1),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.3, 1.0), 
    'feature_fraction': hp.uniform('feature_fraction', .3, 1.0),
    'reg_lambda': hp.uniform('reg_lambda', 0.0, 1.0),
    'reg_alpha': hp.uniform('reg_alpha', 0.0, 1.0),
    'min_split_gain': hp.uniform('min_split_gain', 0.0001, 0.1)
}

space['max_depth']
```

```python
def objective(space):
    model = LGBMClassifier(learning_rate = space['learning_rate'],
                          n_estimators=space['n_estimators'],
                          max_depth=space['max_depth'],
                          num_leaves=space['num_leaves'],
                          colsample_bytree=space['colsample_bytree'],
                          feature_fraction = space['feature_fraction'],
                          reg_lambda=space['reg_lambda'],
                          reg_alpha=space['reg_alpha'],
                          min_split_gain=space['min_split_gain'])
    lc.fit(x_train,y_train)
    y_predict=lc.predict(x_test)
    accuracy = accuracy_score(y_test,y_predict)
    return {'status':STATUS_OK,'loss':-accuracy,'accuracy':accuracy}
```

```python
trials = Trials()
best = fmin(fn=objective,space=space,algo=tpe.suggest,max_evals=150,trials=trials)
best
```

```python
lc = LGBMClassifier(colsample_bytree=best['colsample_bytree'],feature_fraction=best['feature_fraction'],
                    learning_rate=best['learning_rate'],max_depth=int(best['max_depth']),
                   min_split_gain=best['min_split_gain'],n_estimators=int(best['n_estimators']),
                   num_leaves=int(best['num_leaves']),reg_alpha=best['reg_alpha'],reg_lambda=best['reg_lambda'])
lc.fit(x_train,y_train)
y_predict = lc.predict(x_test)
print(accuracy_score(y_test,y_predict))
print(confusion_matrix(y_test,y_predict))
print(classification_report(y_test,y_predict))
```
